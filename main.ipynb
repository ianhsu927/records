```python
import fitz  # PyMuPDF
import pandas as pd

def extract_table_from_pdf(pdf_path, page_number):
    # 打开PDF文件
    document = fitz.open(pdf_path)
    page = document.load_page(page_number)  # 选择要读取的页面
    
    # 获取页面文本
    text = page.get_text("text")
    # 分割文本成行
    lines = text.split('\n')
    
    # 假设每行是用空格或制表符分隔的
    table_data = [line for line in lines]
    table_data = [table_data[i:i+9] for i in range(10, len(table_data), 11)]
    return table_data

# 示例用法
pdf_path = 'records.pdf'  # PDF文件路径
page_number = 0  # 页码，从0开始计数
data = extract_table_from_pdf(pdf_path, page_number)
# data
```
